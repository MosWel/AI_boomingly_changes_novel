```python
import pandas as pd
df = pd.read_csv("novel_sxft.csv")
df.iloc[0,1] = '第零卷 书籍简介'
df.iloc[1,2] = df.iloc[1,2].replace('林风','刘震撼').replace('苏瑶','小狐狸')
```

# 章节名清洗

```python
import re
import numpy as np
pattern1 = re.compile(r"第[一二三四五六七八九十百千零a-z]+章")
pattern2 = re.compile(r"第[一二三四五六七八九十百千零a-z]+卷")
```

## 章节格式规范

```python
idx = []
for dt in range(len(df)):
    if pattern2.match(df.iloc[dt,1]):
        idx.append(df.iloc[dt,0])
        print(df.iloc[dt,1])
        #print(df1.iloc[dt,2])
idx.append(-1)
print(idx)
```

```python
count = 0
tol = 1
for dt in range(len(idx[1:])-1):
    for ddt in df.iloc[idx[dt+1]+1:idx[dt+2], 1]:
        count += 1
        tol += 1
        df.iloc[tol, 1] = re.sub(pattern1, "第{0}章".format(count), ddt)
    count = 0
    tol += 1
del count, tol, dt, ddt
```

```python
## 已经实现对章节名字的清洗，使得数字合理
df.columns = ['index', '前缀', '后缀']
df.set_index('index')

pattern3 = re.compile(r"第[0-9]+章")
for idx, dt in enumerate(df.iloc[:,1]):
    if pattern3.match(dt):
        temp = dt.strip(' ')
        temp = temp.strip('　').split('章', 1)
        df.iloc[idx, 1] = temp[0]+"章 "+temp[1]

del pattern3, idx, dt
```

## 章节名规范

```python
pattern3 = re.compile(r'第[0-9]+章\s')
# pattern5 = re.compile(r'.*[\?]')
pattern6 = re.compile(r'.*\{.+第[0-9]+章\}')
pattern7 = re.compile(r'.*\\[a-z0-9]+')

def chooes(pattern, item):
    if re.match(pattern, item):
        res = re.match(pattern, item)
        print(res)
        return res
# chooes(pattern6, lt[0][idx])

lt = [df['前缀'], df['后缀']]
for idx in range(len(lt[0])):
    if re.match(pattern6, lt[0][idx]):
        print(re.match(pattern6, lt[0][idx]))
    if re.match(pattern7, lt[0][idx]):
        print(re.match(pattern7, lt[0][idx]))
```

# 合并保存

```python
import os

temp_chapter = ''
pattern_os = re.compile(r"第[0-9]+章")

for idx in range(len(lt[0])):
    if pattern2.match(lt[0][idx]):
        temp_chapter = lt[0][idx].split(" ")
        os.makedirs('sxft_data\\{0}'.format(temp_chapter[0]), exist_ok=True)
        with open('sxft_data\\{0}\\本卷简介.txt'.format(temp_chapter[0]), 'w', encoding='ANSI')as _file:
            _file.write(temp_chapter[1]  + "\n\n" + lt[1][idx])
    if pattern3.match(lt[0][idx]):
        os.makedirs('sxft_data\\{0}\\{1}'.format(temp_chapter[0], pattern_os.match(lt[0][idx]).group()), exist_ok=True)
        with open('sxft_data\\{0}\\{1}\\本章内容.txt'.format(temp_chapter[0], pattern_os.match(lt[0][idx]).group()), 'w', encoding='ANSI')as _file:
            _file.write(lt[0][idx].split(" ")[-1] + "\n\n" + lt[1][idx])
    if idx == 780:
        del temp_chapter
        os.makedirs('sxft_data\\{0}'.format('真结局'), exist_ok=True)
        with open('sxft_data\\真结局\\本卷简介.txt', 'w', encoding='ANSI')as _file:
            _file.write(lt[1][idx])
del _file
```

# 检查小说内容长度

```python
import matplotlib.pyplot as plt 

def generate_graph(lt_lengths, title='列表元素', step=10):
    '''
    生成列表元素长度的直方图
    :param lt_lengths: 列表元素长度列表
    :param title: 图表主题
    '''
    plt.hist(lt_lengths, bins=range(0, max(lt_lengths) + 10, step), edgecolor='black')
    plt.title(title + '长度分布')
    plt.xlabel(title + '长度')
    plt.ylabel('统计数量')
    plt.show()

del pattern1, pattern2, pattern3, pattern_os, pattern6, pattern7, temp, idx
```
